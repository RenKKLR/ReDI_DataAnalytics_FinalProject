# Traning and testing set

## ML_Model Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df, test_size=0.2)
```

```python
train_df
```

```python
test_df
```

```python
train_X = train_df.drop('loan_status', axis=1)
train_y = train_df['loan_status']

test_X = test_df.drop('loan_status', axis=1)
test_y = test_df['loan_status']
```

```python
decision = DecisionTreeClassifier()

decision.fit(train_X, train_y)
```

```python
decision.score(test_X, test_y)
```

```python
decision.feature_names_in_
```

```python
decision.feature_importances_
```

```python
importances = dict(zip(decision.feature_names_in_, decision.feature_importances_))
importances = {k: v for k, v in sorted(importances.items(), key=lambda x: x[1], reverse=True)}
importances
```

## ML_Model Random Forest

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
forest = RandomForestClassifier()

forest.fit(train_X, train_y)
```

```python
forest.score(test_X, test_y)
```

## ML_Model Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic = LogisticRegression(max_iter=1000)
logistic.fit(train_X, train_y)
```

```python
logistic.score(test_X, test_y)
```

## Comparision of the 3 models

```python
names = ['DecisionTree', 'RandomForest', 'LogisticRegression' ]
scores = [decision.score(test_X, test_y),forest.score(test_X, test_y),logistic.score(test_X, test_y)]

plt.bar(names, scores)
plt.xlabel('Algorithms')
plt.ylabel('Scores')
plt.title('Comparison of multiple algorithms')
plt.show()
```

## Applying Cross Validation

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
models = []
models.append(('DT', DecisionTreeClassifier()))
models.append(('RF', RandomForestClassifier()))
models.append(('LR', LogisticRegression(max_iter=1000)))
models
```

```python
results = dict()
for name, model in models:
    kfold = KFold(n_splits=10, shuffle=True)
    cv_results = cross_val_score(model, train_X, train_y, cv=kfold, scoring='accuracy')
    results[name]=(cv_results.mean(), cv_results.std())

print("name      results.mean      results.std")
for key, value in results.items():
    print(f"{key:10} {value[0]:.5f}           {value[1]:.5f}")
```
